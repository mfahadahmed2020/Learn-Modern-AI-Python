  #   Types of Inheritance:
  1.  Single Inheritance: A child class inherits from a single  parent class.
  2.  Multiple Inheritance: A child class inherits from more than one parent class.
  3.  Multilevel Inheritance: A child class inherits from a   parent class, which in turn inherits from another class.
  4.  Hierarchical Inheritance: Multiple child classes inherit from a single parent class.
  5.  Hybrid Inheritance: A combination of two or more types of inheritance.  


```python
# Single Inheritance
class Secret_Service:
    def __init__(self, name):
        self.name = name

    def display_name(self):
        print(f"Secret_Service's Name: {self.name}")

class Chief(Secret_Service):  # Single Inheritance
    def sound(self):
        print("Chief X2")

# Multilevel Inheritance
class GuideAgent(Chief):  # Multilevel Inheritance
    def guide(self):
        print(f"{self.name} Assist & Helping The Ali Imran!")

# Multiple Inheritance
class SecretService:
    def greet(self):
        print("SecretService!")

class GoldenRetriever(Secret_Service, SecretService):  # Multiple Inheritance
    def sound(self):
        print("Ali Imran Had Kept Tahir As His Dummy AS AN Chief - X2!")

# Example Usage
lab = Chief("Ali Imran")
lab.display_name()
lab.sound()

guide_assist = GuideAgent("Tahir")
guide_assist.display_name()
guide_assist.guide()

retriever = GoldenRetriever("Safdar Is Members")
retriever.display_name()
retriever.greet()
retriever.sound()
```

```python
# Parent Class
class X2:
    def voice(self):
        print("Chief X2")  # Default implementation

# Run-Time Polymorphism: Method Overriding
class Chief(X2):
    def voice(self):
        print("Tahir Dummy X2 ")  # Overriding Parent Method

class AliImran(X2):
    def voice(self):
        print("Ali Imran is Real X2")  # Overriding Parent Method

# Compile-Time Polymorphism: Method Overloading Mimic
class Ranculator:
    def grade(self, a, b=0, c=0):
        return a + b + c  # Supports Multiple Ways To Call Add()

# Run-Time Polymorphism
chief = [X2(), Chief(), AliImran()]
for chief in chief:
    chief.voice()  # Calls The Appropriate Method Based On The Object Type


# Compile-Time Polymorphism (Mimicked Using Default Arguments)
rank = Ranculator()
print(rank.grade(5, 10))  # Two Arguments
print(rank.grade(5, 10, 15))  # Three Arguments
```

    Explanation:

    Python does not natively support method overloading. Instead, we use a single method (add) with default arguments to handle varying numbers of parameters.

 *3. Python Encapsulation
    Encapsulation is the bundling of data (attributes) and methods (functions) within a class, restricting access to some components to control interactions.

    A class is an example of encapsulation as it encapsulates all the data that is member functions, variables, etc.

    Methods

    Types of Encapsulation:
    
    Public Members: Accessible from anywhere.
    
    Protected Members: Accessible within the class and its subclasses.
    
    Private Members: Accessible only within the class.

```python
class Chief:
    def __init__(self, name, secret, rank):
        self.name = name  # Public Attribute
        self._secret = secret  # Protected Attribute
        self.__rank = rank  # Private Attribute

    # Public Method
    def get_info(self):
        return f"Name: {self.name}, Secret: {self._secret}, Rank: {self.__rank}"

    # Getter And Setter For Private Attribute
    def get_rank(self):
        return self.__rank

    def set_rank(self, rank):
        if rank > 0:
            self.__rank = rank
        else:
            print("Tahir As Nick-Code Black Zero!")

# Example Usage
chief = Chief("Ali Imran Publicaly", "Tahir", "X2")

# Accessing Public Member
print(chief.name)  # Accessible

# Accessing Protected Member
print(chief._secret)  # Accessible but discouraged outside the class

# Accessing Private Member Using Getter
print(chief.get_rank())

# Modifying Private Member Using Setter
chief.set_rank(0)
print(chief.get_info())
```

Explanation:

  * Public Members: Easily Accessible, Such As Name.
  
  * Protected Members: Used With A Single _, Such As _Secret. Access is  
    Discouraged but allowed in subclasses.
  
  * Private Members: Used With __, Such As __Rank. Access Requires Getter And
    Setter Methods.

  * 4. # Data Abstraction
        
        Abstraction Hides The Internal Implementation Details While Exposing Only The Necessary Functionality. It Helps Focus On "What To Do" Rather Than "How To Do It."

  *     Types Of Abstraction:
        
  *      Partial Abstraction: Abstract Class Contains Both Abstract
         And Concrete Methods.

  *      Full Abstraction: Abstract Class Contains Only Abstract  
         Methods (Like Interfaces).

```python
from abc import ABC, abstractmethod

class Chief(ABC):  # Abstract Class
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def voice(self):  # Abstract Method
        pass

    def display_name(self):  # Concrete Method
        print(f"First & Second Chief Name: {self.name}")

class Helper(Chief):  # Partial Abstraction
    def voice(self):
        print("Chief - X2!")

class Assist(Chief):  # Partial Abstraction
    def voice(self):
        print("Voice of X2!")

# Example Usage
chief = [Helper("Ali Imran"), Assist("Tahir")]
for chief in chief:
    chief.display_name()  # Calls Concrete Method
    chief.voice()  # Calls Implemented Abstract Method
```

Explanation:

  Partial Abstraction: The Chief Class Has Both Abstract (Voice) And Concrete (Display_Name) Methods.

  Why Use It: Abstraction Ensures Consistency in Derived Classes By Enforcing The Implementation of Abstract Methods

```python
# Your original base
class Person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def speak(self):
        print(f"{self.name} says hello!")

    def work(self):
        print("Person is working")

# Two concrete subclasses
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def study(self):
        print(f"{self.name} is studying")

    def work(self):
        print("student is studying")

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def teach(self):
        print(f"{self.name} is teaching {self.subject}")

    def work(self):
        # overrides Person.work
        print(f"{self.name} is preparing lessons for {self.subject}")

# New class with multiple inheritance
class TeachingAssistant(Student, Teacher):
    def __init__(self, name, age, student_id,subject):
       # Manually initialize Person once
       Person.__init__(self, name, age)
       # Then set subclass-specific data
       self.student_id = student_id
       self.subject = subject

    # We get both study() and teach() “for free”
    # Optionally override work() to combine behaviors:
    def work(self):
        print(f"{self.name} is grading assignments and helping in {self.subject}")

# Usage
ta = TeachingAssistant("Ali", 25, "5456", "Math")
ta.speak()
ta.work()
# ta = TeachingAssistant("Aisha", 25, "S456", "Math")
# ta.speak()         # from Person
# ta.study()         # from Student
# ta.teach()         # from Teacher
# ta.work()          # overridden in TA
print(TeachingAssistant.mro())
```

```python
class Person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def work(self):
        print("person is working")

    # String representation for print()
    def __str__(self):
        return f"{self.name}, age {self.age}"

    # Overload '+' to merge two Person instances
    def __add__(self, other):
        if not isinstance(other, Person):
            return NotImplemented
        # Combine names and sum ages
        combined_name = f"{self.name} & {other.name}"
        combined_age = self.age + other.age
        return Person(combined_name, combined_age)

    # Overload '==' to compare name and age
    def __eq__(self, other):
        if not isinstance(other, Person):
            return NotImplemented
        return self.name == other.name and self.age == other.age


class Teacher(Person):
    def work(self):
        print("teacher is teaching")


class Student(Person):
    def work(self):
        print("student is studying")


# Demonstration
p1 = Person("Ali", 20)
p2 = Person("Ayesha", 22)
p3 = p1 + p2
print(p3)            # Ali & Ayesha, age 42

# Equality check
p4 = Person("Ali", 20)
print(p1 == p4)      # True
print(p1 == p2)      # False

# Polymorphic work() calls
for obj in (p1, Teacher("Abdullah", 30), Student("Aamir", 18)):
    obj.work()
```

```python
# class / instance lever atrributes

class Person():
  population = 0
  default_language = "Urdu"

  def __init__(self, name, age, langauge:str = None ):
    Person.population += 1
    self.name = name
    self.age = age
    self.language = langauge if langauge else Person.default_language


person1 = Person("Ali",20, "english")
person2 = Person("Abdullah",25)
print("population",Person.population)
print("person 1 langauge",person1.language)
print("person 2 language",person2.language)
# print("default langauge",Person.default_language)
```
