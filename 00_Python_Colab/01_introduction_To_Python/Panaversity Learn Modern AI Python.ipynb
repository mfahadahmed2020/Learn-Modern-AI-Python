- Python ŸÖ€å⁄∫ ÿ¢Ÿæÿ±€åŸπÿ±ÿ≤ ⁄©Ÿà ÿ≠ÿ≥ÿ® ÿ∂ÿ±Ÿàÿ±ÿ™ ÿ¢ÿ®ÿ¨€å⁄©Ÿπÿ≥ ⁄©€í ŸÑ€å€í ÿßŸàŸàÿ± ŸÑŸà⁄à ⁄©€åÿß ÿ¨ÿß ÿ≥⁄©ÿ™ÿß €Å€í€î
- + (addition) ÿßŸàÿ± ** (exponentiation)** ÿØŸàŸÜŸà⁄∫ ⁄©Ÿà ÿßŸàŸàÿ± ŸÑŸà⁄à ⁄©€åÿß ÿ¨ÿß ÿ≥⁄©ÿ™ÿß €Å€í ÿ™ÿß⁄©€Å ÿ¢Ÿæ ÿßŸæŸÜ€í ⁄©ŸÑÿßÿ≥ÿ≤ ⁄©€í ŸÑ€å€í ÿßŸÜ ÿ¢Ÿæÿ±€åŸπÿ±ÿ≤ ⁄©€å ŸÖÿÆÿ™ŸÑŸÅ ⁄©ÿßÿ±⁄©ÿ±ÿØ⁄Ø€å ŸÅÿ±ÿß€ÅŸÖ ⁄©ÿ± ÿ≥⁄©€å⁄∫€î
- ÿ¢Ÿæ _add_ ÿßŸàÿ± _pow_ ÿ¨€åÿ≥€å ÿÆÿµŸàÿµ€å ŸÖ€åÿ™⁄æ⁄àÿ≤ ⁄©Ÿà ÿßŸæŸÜ€å ⁄©ŸÑÿßÿ≥ÿ≤ ŸÖ€å⁄∫ ⁄àŸÅÿßÿ¶ŸÜ ⁄©ÿ± ⁄©€í €å€Å ÿ¢Ÿæÿ±€åŸπÿ±ÿ≤ ÿßŸàŸàÿ± ŸÑŸà⁄à ⁄©ÿ± ÿ≥⁄©ÿ™€í €Å€å⁄∫€î

```python
# Lesson 01 To 09 - Grand Test No :11
class MyNumber:
    def __init__(self, value): # Changed _init_ to __init__
        self.value = value

    def __add__(self, other): # Changed _add_ to __add__
        return MyNumber(self.value + other.value)

    def __pow__(self, power): # Changed _pow_ to __pow__
        return MyNumber(self.value ** power)

num1 = MyNumber(3)
num2 = MyNumber(4)

# Adding two MyNumber objects using overloaded '+' operator
result = num1 + num2  # Calls __add__()
print(result.value)  # Output: 7
```
