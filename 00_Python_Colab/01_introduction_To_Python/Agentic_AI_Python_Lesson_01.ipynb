1. پائیتھون کے مقبول ہونے کی ایک اہم وجہ کیا ہے؟

  The correct answer is:

---

Explanation:

One of the key reasons Python is so popular is because of its easy-to-read and simple syntax, which is similar to the English language. This makes it beginner-friendly and highly productive for developers of all levels.

Other reasons include:
- Vast library support
- Large and active community
- Versatility (web dev, data science, AI, etc.)

So, options A, B, and D are incorrect.
