8. `_call_`, Meta Use Preview (Optional Advanced)

Class Meta

````python
class Meta(type):
  def __call__(cls, *args, **kwargs):
    print("[Meta] Call Used During Object Creation.")
    return super().__call__(*args, **kwargs)

class MyClass(metaclass=Meta):
  def __init__(self):
    print("[MyClass] __init__ Executed.")

m = MyClass()
````

Object Internals - Output All Internal Methods/Attributes

````python
class Spy:
    def __new__(cls, *args, **kwargs):
        print(">> __New__IS Called")
        # Correctly Call The SuperClass's __New__Method
        instance = super().__new__(cls)
        return instance

    def __init__(self, name, code):
        print(">> __init__ IS Called")
        self.name = name
        self.code = code

    def __str__(self):
        return f"SPY({self.name}, {self.code})"

    def __repr__(self):
        return f"<SPY Name={self.name} Code={self.code}>"

# Create Object
agent = Spy("Ali Imran", "X2")

#Background Outputs
print("\n--- Object Internals ---")
print("Type:", type(agent))
# The Attribute __Class__IS Correct
print("Class:", agent.__class__)
# The Attribute__Dict__IS Correct
print("Dict:", agent.__dict__)
# Calling STR() AND REPR() Will Implicitly Use __STR__AND__REPR__
print("STR:", str(agent))
print("REPR:", repr(agent))
# The Method __Size Of__IS Correct
print("SIZE:", agent.__sizeof__(), "BYTES")
# Dir() IS Correct
print("DIR (All Methods):", dir(agent))
````
